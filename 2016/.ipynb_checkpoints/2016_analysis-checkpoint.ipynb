### Part a)

- For J large and negative, we would expect to see a disorded phase at high T, where the entropic free energy contribution causes random ordering of spins. At low T, we would expect to see an anti-ferromagnetic phase, where spins are ordered in an alternating up-down-up pattern. This minimises the enthalpic free energy contribution.
- For J=-1, h large and positive, we would again expect a disordered phase at high T, but at low temperature we would expect to see ferromagnetic-type alignment of spins, as up-spins (+1) minimise the external field term in the expression for the total system energy.

### Part b) 
The code required for simulation of the antiferromagnet Ising model is held in two files:
- eqns.py contains the required functions.
- vis.py shows a real-time simulation of the evolving system over time. The desired value of h is taken as an additional command line argument (i.e run 'python3 vis.py h').
