## Part d) - computing variance

```python
df_d=pd.read_excel('variance.xlsx')
```

```python
df_d.head()
```

```python
x=df_d['p'].to_numpy()
y=df_d['variance'].to_numpy()
yerr=df_d['variance error'].to_numpy()

plt.title('Plot of variance as a function of p')
plt.ylabel('Variance in number of infected sites')
plt.xlabel('Probability, p')
plt.errorbar(x, y, yerr=yerr)
```
