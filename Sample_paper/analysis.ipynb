```python
df=pd.read_excel('phi.xlsx') #import phi data
```

```python
df=df.drop(('Unnamed: 0'), axis=1) #clean up data
```

```python
phi=df.to_numpy()
```

```python
#generate NxN array of radial distances

R=np.zeros((50,50))

for i in range(50):
    for j in range(50):

        c=int(50/2)

        x=c-i
        y=c-j

        r_sq=x**2+y**2

        R[i,j]=np.sqrt(r_sq)
        
```

```python
phi_flat=phi.flatten()
R_flat=R.flatten()
```

```python
plt.title('Phi vs radial distance')
plt.ylabel('Phi')
plt.xlabel('r')
plt.scatter(R_flat[0:1250],phi_flat[0:1250])
plt.show()
```

```python
#define exponential and x^n functions

def exp(r,C,a):
    """
    Returns exponential of radial distance
    """
    phi=C*np.exp(a*r)
    
    return phi

def power(r,C,n):
    """
    Returns radial distance raised to some power, n
    """
    phi=C*np.power(r, n)
    
    return phi
```

```python
popt=curve_fit(power, R_flat, phi_flat)[0]
print(popt)
```

Fitting to a power law provides the lowest covariance betweem fitted and simulated data. The optimised power law is:

## Introducing spatially dependent velocity

The case where the field $\phi$ is advected by a spatially dependent velocity $\vec{v}$ is shown below. Going from left to right, the cases $v_\circ=0.01$, $v_\circ=0.1$ and $v_\circ=0.5$ are shown.

The contour plot of $\phi$ is slightly skewed by the new velocity term, appearing more intense towards the left(low y-coordinates) than on the right (large y-coordinates). This can be interpreted as the velocity gradient, which increases from left to right increasing the concentration at lower velocities, where diffusion is less prominent.
